### 3. Data Types in Pandas:
It has two data types:
1. Series --> they are 1-Dimensional Data Type. 
2. DataFrames --> they are 2-D and use python dictionaries.

#### I. Series
`Series()` function is used to make series in Pandas and this functions take one-Dimentional data structures (like lists).
Below is the example.

```python
car_names=pd.Series(['BMW','HONDA','TOYOTA','COROLA'])
car_names
```

```python
colors=['blue','metalic','grey','white']
car_colors=pd.Series(colors)
car_colors 
```

#### II. DataFrames:
`DataFrame()` is the function which is being used for making pandas DataFrames. It takes 2-dimentional data structues and make a dataframe of them.
Below is an example and Python Dictionary which is 2 dimensional structure is given to it.
Name of the key in dictionary refers to the Column Name in dataframe and value in dictionary refers to the items in the data frame rows and columns.

```python
cars_detail=pd.DataFrame({'car name':car_names,'car color':car_colors})
cars_detail
```

`car name` and `car color` are the series which we have created in previous cells. We can pass some lists here too.

```python
cars_detail=pd.DataFrame({'car name':["BMW","FORD"],'car color':["Black","Blue"]})
cars_detail
```

Instead of creating DataFrames with hands, we can also import data to pandas dataframe.  

#### Reading and Exporting CSV file using pandas:
`read_csv()` function does this magic for us. It takes the path of csv file and then convert data into dataframe.

```python
sales=pd.read_csv("100 Sales Records.csv")
sales
```

And in the same manner we can also convert/export our dataframe into `csv` file using the `to_csv()' function by passing it the path where to store the file.

```python
cars_detail.to_csv("cars_details",index=False) #index=False indicates that it will be saved without and indexing.
```

### 5. Selecting and Viewing data:

`head()` is used to see from above side of dataframe and you can tell that how many rows you wanna see.

```python
sales.head(4)
```

`tail()` is used to see from below side of dataframe and you can tell that how many rows you wanna see.

```python
sales.tail(3)
```

`loc[]` shows the information of the item at that index.

```python
sales.loc[99]
```

`iloc[]` tells the position of the item passad and also enables the slicing

```python
sales.iloc[3]
```

```python
sales.iloc[: : -1] #it is just like python lists slicing and it displays DataFrame in reverse.
```

#### How to access a column?
Here are the two ways.

```python
#1. by passing column name
sales['Country']
```

```python
#2. we can acces columns with '.' notation. `dataframe.columnname`
sales.Country
```

#### Columns Operations:

```python
#shows data items of dataframes where Item type is Clothes
sales[sales["Item Type"]=="Clothes"]
```

```python
#5.groupby shows one column in according/comparison to other. 
sales['Item Type'].groupby(sales["Country"]).sum()
```

```python
#6crosstab: advanced view of cross tab
pd.crosstab(sales['Country'],sales['Units Sold'])
```

### 6. Basic Visualization of Data using Matplotlib:

### 7. Data Manupulating:
Here are the few functions to do data manipulation.

```python
#str.lower(): converts the string value to lower case.
sales.Country=sales["Country"].str.lower()
```

```python
sales.Country.head(10) #seeing change
```

```python
sales.isna() #shows all the missing values in data,false=not missing,true=missing
```

```python
#filling a missing value in column: using `fillna()`.
sales['Country'].fillna('US')
```

```python
#dropna():drop all missing values from DataFrame.
sales['Country'].dropna(inplace=True)#inplace refers to change in actual dataframe
```

```python
#drop(): drops a column from dataframe
sales.drop("Region",axis=1,inplace=True)
```

```python
#viewing the changes
sales.head() #see there is no region
```

```python
#Taking a sample from data 
sales_sample=sales.sample(frac=0.01) #fraction defines the size of sample
```

```python
sales_sample
```

```python
#reseting index
sales.reset_index()
```
